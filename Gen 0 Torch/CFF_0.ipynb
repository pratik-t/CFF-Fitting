```python
import torch
import numpy as np
%matplotlib widget
import matplotlib.pyplot as plt
from bkm10 import BKM10
```

```python
k = 5.75
Q2 = 1.82
x = 0.343
t = -0.172
phi = np.pi/2
kins = torch.Tensor([[k, Q2, x, t, phi]])

ReH = -2.56464
ReE = 2.21195
ReHt = 1.39564
dvcs = 0.0315875


cffs = torch.Tensor([[ReH, ReHt, ReE, dvcs]])

compute = BKM10()
dsig = compute.diff_cross(kins, cffs, helc = 0)
print(dsig)
```
